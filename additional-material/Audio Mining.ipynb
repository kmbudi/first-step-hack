```python
#Import library
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

from glob import glob
#!pip install librosa
import librosa as lr    #mengubah audio stereo menjadi mono
import librosa.display  

import IPython.display as ipd   #memunculkan suara pada python
#import nir_evaL

from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
#set directory for sources file
#data dapat diunduh pada link: https://www.kaggle.com/datasets/juliancienfuegos/what-is-a-note/code
data_dir = './note/GuitarNotes'
audio_files = glob(data_dir + '/*.wav')     #/*.wav -> memanggil keseluruhan file audio pada folder
len(audio_files)
```

# Read and Visualize

```python
#read first audio file, set time array
audio, sfreq = lr.load(audio_files[0])
time = np.arange(0, len(audio))/sfreq
time
```

```python
#plot audio over time
fig, ax = plt.subplots()
ax.plot(time,audio)
ax.set(xlabel = 'Time (s)', ylabel = 'Sound Amplitude')
plt.show()
```

```python
#read all audio files
for file in range(0, len(audio_files), 1):
    #read audio file, set time array
    audio, sfreq = lr.load(audio_files[file])
    time = np.arange(0, len(audio))/sfreq
    
    #plot audio over time
    fig, ax = plt.subplots()
    ax.plot(time,audio)
    ax.set(xlabel = 'Time (s)', ylabel = 'Sound Amplitude')
    plt.show()
```

# Spectogram

```python
#Spektogram file audio 1
y, sr = lr.load(audio_files[0])
D = lr.stft(y)
S_db = lr.amplitude_to_db(np.abs(D), ref=np.max)
print(S_db.shape)

# Plot the transformed audio data
fig, ax = plt.subplots(figsize=(10, 5))
img = lr.display.specshow(S_db,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Spectogram', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

# Mel Spectogram

```python
S = lr.feature.melspectrogram(y=y,
                                   sr=sr,
                                   n_mels=128 * 2,)
S_db_mel = lr.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the mel spectogram
img = lr.display.specshow(S_db_mel,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Mel Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```
